# 1.6 Arrays Concatenation:

* NumPy allows you to join a sequence of arrays along a specified and strictly existing axis. for example, concatenating one dimensional arrays `a` and 'b' along axis 0 results in a new one dimensional array having the same values as `a` and `b` along axis 0. it is important to note that those arrays cannot be concatenated along axis 1 since both arrays `a` and `b` do not have axis 1. this is the syntax of the NumPy functionality of concatenation. the first input argument is a tuple of arrays that you wish to concatenate. the second input argument is the axis along which you want to concatenate those arrays. by default, axis value is set to zero, but you can set it to any axis value. Screen Shot 2022-06-12 at 5.54.48 PM.png


* in this case, we have arrays `a` and `b` each of which is a two dimensional array. since a two dimensional array has axes 0 and 1, both `a` and `b` can be concatenated either along axis 0 or axis 1. when `a` and `b` and are concatenated along axis 0, the concatenate function will concatenate `a` and `b` along the definition of zero axis of a two dimensional array. therefore, both `a` and `b` will be concatenated row-wise. similarly, when concatenating along axis 1, the concatenate function will concatenate both `a` and `b` according to location of axis 1 in the definition of a two dimensional array. therefore, both arrays `a` and `b` are concatenated column-wise. # 1.7 Arrays Stacking:

* NumPy allows you to join multiple arrays using not only concatenation method, but also the stacking method. one of the main differences between stacking and concatenation is that the axis at which concatenation is applied must exist in all arrays to be concatenated. however, stacking allows you to stack in an axis that does not necessarily exist in all arrays to be stacked.

  ```python
  np.stack((a,b,..),axis=0)
  ```



* as the name suggest, stack methods refers to stacking arrays upon each other at a specified axis. when stacking two or more arrays, the dimensionality of the stacked arrays will increase by 1. for example, stacking the one dimensional arrays `a` and `b` will yield a two dimensional array. furthermore, the axis value that you assign in the stack method is the location of the axis of the resulting stacked array not the original arrays. in other words, stacking happens at the axis in the result array along which the input arrays are stacked. for example, in the first case, stacking along axis 0 will allow you to stack the arrays along axis 0 where it is located in the resulting two dimensional array, so arrays `a` and `b` are stacked horizontally. however, stacking arrays along axis 1 in the resulting two dimensional arrays will vertically stack arrays `a` and `b`. notice how values of each array `a` and `b` are re-arranged vertically to match the specified stacking axis. now, let’s look at stacking two dimensional arrays.

# 1.11 Other Methods to Create NumPy Arrays:

* what we used so far to create NumPy array is the `np.array()` to convert other data types to NumPy arrays such as Python lists. However, there are more techniques that create NumPy arrays. in this section, those functions functions will be discussed:

## 1. `np.arange()`:



* `np.arange()` returns evenly spaced values within a given interval. This is the syntax of of the `np.arange()`. 

  ```python
  numpy.arange([start, ]stop, [step, ])
  ```
  *start* is an inclusive starting value in the created array. *stop* is an execlusive ending value in the array. *step* is the spacing between values.  since *start* and *step* are in brackets, it means they are optional. the default value of *start* is 0 and and the default value for *step* is 1. 


* in this example, an array whose starting value is 0 and ending value of 24 is created. in this case, 24 is execlusive. notice that *start* and *step* were not specified. 

````python
x = np.arange(24)
print(x.shape)
print(x)
````

* you can also reshape it to a two dimensional array with shape `(4,4)` for example.

````python
x = np.arange(24).reshape(4,6)
print(x.shape)
print(x)
````

## 2. `np.zeros()` and `np.ones()`:



* `np.zeros()` is used to create an array full of 0 values, and `np.ones() is used to create an array full of 1 values`. this is the syntax for both.
```python
np.zeros(shape_tuple)
np.ones(shape_tuple)
``` 
both of them take the required output shape with either zeros or ones values.


````python
x = np.zeros((3,6)) # creates an array full of zeros with shape (3,6)
print(x.shape)
print(x)
````

````python
x = np.ones((5,3)) # creates an array full of ones with shape (5,3)
print(x.shape)
print(x)
````

## 3. `np.eye()`:


* it returns a two dimensional array with ones on the diagonal and zeros elsewhere.
```python
np.eye(N, M)
```

  *N* is the number of rows in the created array, and *M* is the is the number of columns in the created array. *M* has default value of *N*. 



````python
x = np.eye(4) # creates a (4,4) eye array
print(x.shape)
print(x)
````

````python
x = np.eye(4,2) # creates a (4,2) eye array
print(x.shape)
print(x)
````

## 4. `np.linspace()`:


* similar to `np.arange()`, the `np.linspace()` returns evenly spaced numbers over a specified interval. However, the main difference between `np.arange()` and `np.linspace()` is that `np.arange()` allowes you to set step size, but the `np.linspace()` allows you to define the number of steps between *start* and *end* values. here is the syntax of the `np.linspace()`:

  ```python
  numpy.linspace(start, end, num=50, endpoint=True)
  ```
  * *start* is starting value in the created array.
  * *end* is the ending value in the created array.
  * *num* is the number of data points between *start* and *end*. the default value of the *num* is 50.
  * if *endpoint* is `True`, then *end* is inclusive, execlusive otherwise. The defualt value for *endpoint* is `True`.

````python
x = np.linspace(0,20) # creates an array starting from 0 ending at 20 with 50 datapoints in between
print(len(x),'\n')
print(x)
````

````python
x = np.linspace(0,20,5) # creates an array starting from 0 ending at 20 with 5 datapoints in between
print(len(x),'\n')
print(x)
````

````python
x = np.linspace(0,20,5,endpoint=False) # creates an array starting from 0 ending at 20 with 5 datapoints in between. end is execlusive.
print(len(x),'\n')
print(x)
````

# 1.18 Numerical Derivatives and Integrations:

* using NumPy, you can compute the numerical derivative of an array using the `np.gradient()` function. 

````python
x = np.arange(0, 10, 0.5) # create a range of values from 0 (inclusive) to 10 (execlusive) with steps of 0.5

f = 0.1*np.exp(x - 1) # numpy exponential function with a factor of 0.1 and shifted by 1

df_dx = np.gradient(f,x, axis=0) # derivative of f with respect to x. axis has None value by default, so you can set derivative along any axis you want.

print('x = ',x,'\n')
print('df/dx = ',df_dx)
````

* you can also compute the numerical integration of an array using the cumulative summation in NumPy `np.cumsum()`. To compute the integral, you need to multiply the cumulative sum of array `f` by the spacing of the axis the it was calculated upon `x`. The spacing is the one used in the `np.arange()` to create spacing between values of `x`.
  ```python
  numerical_integral = np.cumsum(f)*dx
  ```

````python
dx = 0.5 # step size for variable x

x = np.arange(0, 10, dx) # create a range of values from 0 (inclusive) to 10 (execlusive) with steps of 0.5

f = 0.1*np.exp(x - 1) # numpy exponential function with a factor of 0.1 and shifted by 1

f_integral = np.cumsum(f, axis=0) * dx # integration of f. axis has None value by default, so you can set cumsum along any axis you want.

print('x = ',x,'\n')
print('f_integral = ',f_integral)
````
